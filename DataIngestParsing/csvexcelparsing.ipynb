```python
import pandas as pd
import os
```

```python
os.makedirs("data/structured_files",exist_ok=True)
```

```python
data = {
    'Product' : ['Laptop' , 'Monitor'],
    'Category' : ['Electronics' , 'Accessories'],
    'Price' : [99, 500],
    'Stock' : [2,3],
    'Description' : ["high performance" , "beautiful"]
}

df = pd.DataFrame(data)
df.to_csv("data/structured_files/products.csv",index=False)
```

```python
with pd.ExcelWriter('data/structured_files/products.xlsx') as writer :
    df.to_excel(writer,sheet_name="Products",index=False)
    pd.DataFrame(data).to_excel(writer,sheet_name="Summary",index=False)
```

```python
from langchain_community.document_loaders import CSVLoader,UnstructuredCSVLoader
```

```python
csv_loader = CSVLoader(file_path= "data/structured_files/products.csv",encoding='utf-8',csv_args={'delimiter' : "," , 'quotechar' : '"'})

csv_docs = csv_loader.load() 
print(f"length of document is {len(csv_docs)}") 
print(f"first page of the doc is {csv_docs[0].page_content}")
```
