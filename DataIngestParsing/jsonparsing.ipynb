```python
from langchain_community.document_loaders import JSONLoader
import json
```

```python
event_loader = JSONLoader(
    file_path="data/jsonfiles/timestamp_data.jsonl",
    jq_schema='.events[]',
    text_content=False
)

event_docs = event_loader.load() 
print(f"length of document is {len(event_docs)}") 
print(f"first page of the doc is {event_docs[0].page_content}")
```
