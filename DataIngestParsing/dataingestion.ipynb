```python
from langchain_community.document_loaders import TextLoader #single file
loader = TextLoader("data/text_files/intro.txt",encoding = 'utf-8')
documents = loader.load()
print(documents)
```

```python
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "data/text_files",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={'encoding' : 'utf-8'},
    show_progress=True
)

documents = dir_loader.load()
```

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)

print(documents)
```

```python
text = documents[0].page_content
print(text)
```

```python
# # character based text splitter
char_splitter = CharacterTextSplitter(
    separator='\n' ,
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len
)
character_chunks = char_splitter.split_text(text)

print(f"Created {len(character_chunks)} chunks")
print(f"First chunk is {character_chunks[0][:100]}...")
```
