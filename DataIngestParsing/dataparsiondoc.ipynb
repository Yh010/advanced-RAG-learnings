```python
from docx import Document as DocxDocument
import os
```

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
## using docx2txtloader 
try:
    docx_loader = Docx2txtLoader("data/word/AI_Research_Agent.docx")
    docs = docx_loader.load()
    print(f"length of document is {len(docs)}")
    print(f"first page of the doc is {docs[0].page_content[:200]}...")

except Exception as e:
    print(f"error: {e}")
```

```python
## using UnstructuredWordDocumentLoader
try:
    unstructured_docx_loader = UnstructuredWordDocumentLoader("data/word/AI_Research_Agent.docx")
    unstructured_docs = unstructured_docx_loader.load()
    print(f"length of document is {len(unstructured_docs)}")
    for i,doc in enumerate(unstructured_docs[:3]):
        print(f"element number {i+1}")
        print(f"content is {doc.page_content[:200]}...")

except Exception as e:
    print(f"error: {e}")
```
